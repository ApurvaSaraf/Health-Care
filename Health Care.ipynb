```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data=scaler.fit_transform(X)
data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data,y,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression()
```

```python
model=Lr.fit(X_train,y_train)
```

```python
y_pre=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_pre,y_test))
print(accuracy_score(y_pre,y_test)*100)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pr=clf.predict(X_test)
print(confusion_matrix(y_pr,y_test))
print(accuracy_score(y_pr,y_test)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train,y_train)
y_pr=clf.predict(X_test)
print(confusion_matrix(y_pr,y_test))
print(accuracy_score(y_pr,y_test)*100)
```
